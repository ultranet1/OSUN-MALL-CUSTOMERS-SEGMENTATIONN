# EXPLORATORY DATA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('/storage/emulated/0/Android/data/ru.iiec.pydroid3/files/OSM DC.csv')
print (data)
data=data.drop('Unnamed: 0', axis=1)
print(data)
                 
```

```python
print(data.describe().round(2))
```

```python
sns.countplot(data['Gender'])
plt.legend('male')
plt.title('Gender count')
plt.show()
sns.countplot(data['Age'])
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=20))
plt.show()
sns.countplot(data['Monthly Income per thousand (Naira)'])
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=20))
plt.show()
```

```python
sns.distplot(data['Age'], hist=True)

plt.show()
sns.distplot(data['Monthly Income per thousand (Naira)'], hist=True)
plt.show()
sns.distplot(data['Spending Score (1-100)'], hist=True)
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(data['Spending Score (1-100)'], data['Monthly Income per thousand (Naira)'], hue=data['Gender'] )  
```

```python
sns.scatterplot(x=data['Monthly Income per thousand (Naira)'], y=data['Spending Score (1-100)'], hue=data['Age'])
```
